```python
# Import PyTorch
import torch

# Create tensors
a = torch.tensor([[1, 2],
                  [3, 4]])

b = torch.tensor([[5, 6],
                  [7, 8]])

print("Tensor A:\n", a)
print("Tensor B:\n", b)
print("\nAddition:\n", a + b)
print("\nSubtraction:\n", a - b)
print("\nMultiplication:\n", a * b)
print("\nScalar Multiplication:\n", a * 2)
print("\nMatrix Multiplication:\n", torch.matmul(a, b))
```
