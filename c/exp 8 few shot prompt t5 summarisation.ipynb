```python
# Install libraries
!pip install transformers sentencepiece -q

# Import libraries
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load FLAN-T5 model
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small")

# One-shot example
example = """
Dialogue:
Riya: Hi Rahul, are you coming to college tomorrow?
Rahul: Yes, we have the AI practical exam.
Riya: Don't forget to bring the assignment.
Rahul: Thanks for reminding me.

Summary:
Riya reminds Rahul about the AI practical and assignment.
"""

# New dialogue
dialogue = """
Dialogue:
Amit: Hello Sneha, did you complete the Python project?
Sneha: Yes, I finished it yesterday.
Amit: Great! Can you help me with the final report?
Sneha: Sure, I will send you the format tonight.

Summary:
"""

# Final prompt
prompt = example + dialogue

# Convert text into tokens
inputs = tokenizer(prompt, return_tensors="pt")

# Generate summary
outputs = model.generate(
    inputs["input_ids"],
    max_length=50
)

# Convert output into text
summary = tokenizer.decode(
    outputs[0],
    skip_special_tokens=True
)

# Print summary
print("MODEL SUMMARY:\n")
print(summary)
```
