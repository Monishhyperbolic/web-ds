```python

#Responsible AI=Fairness+Transparency+Safety
# Install libraries
!pip install transformers torch

# Import pipeline
from transformers import pipeline

# Load BERT model
unmasker = pipeline("fill-mask", model="bert-base-uncased")

# Test sentences
sentences = [
    "The doctor said that [MASK].",
    "The nurse said that [MASK]."
]

# Check predictions
for s in sentences:

    print("\nSentence:", s)

    result = unmasker(s)

    # Print top prediction
    print("Prediction:", result[0]['token_str'])
```

```python
from transformers import pipeline

# Load sentiment model
classifier = pipeline("sentiment-analysis")

# Similar sentences
sentences = [
    "The man is intelligent.",
    "The woman is intelligent."
]

# Check outputs
for s in sentences:
    print(s, "->", classifier(s))
```
