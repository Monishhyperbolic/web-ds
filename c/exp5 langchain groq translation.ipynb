```python
# Install libraries
!pip install langchain langchain-groq -q

# Import libraries
import os
from langchain_groq import ChatGroq

# Set API key
os.environ["GROQ_API_KEY"] = "your_api_key"

# Load model
llm = ChatGroq(
    model="llama-3.3-70b-versatile"
)

# Chat loop
while True:

    # User input
    text = input("You: ")

    # Stop chatbot
    if text == "exit":
        break

    # Get AI response
    response = llm.invoke(text)

    # Print response
    print("Bot:", response.content)
```
